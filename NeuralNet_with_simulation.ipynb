```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPRegressor

# Simulate data
np.random.seed(42)
X = np.linspace(0, 10, 1000)
Y = 4 + 3 * X - 0.7 * X**2 + 0.05 * X**3 + np.random.normal(scale=1.5, size=X.shape)  # Quadratic relationship with noise

# Define the MLPRegressor with a single hidden layer of 3 neurons and ReLU activation function
mlp = MLPRegressor(hidden_layer_sizes=(3,), activation='relu', alpha=0, max_iter=1000000, solver='adam', learning_rate='adaptive')

# Fit the model to the data
mlp.fit(X.reshape(-1, 1), Y)

# Predict Y using the fitted model
Y_pred = mlp.predict(X.reshape(-1, 1))

# Plot the results
plt.figure(figsize=(10, 6))
plt.scatter(X, Y, color='blue', label='Data')
plt.plot(X, Y_pred, color='red', label='Neural Net', linewidth=2)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatterplot with Neural Net Predictions')
plt.legend()
plt.show()

# View the weights (coefs) of the model
print("Weights between input and hidden layer:", mlp.coefs_[0])
print("Weights between first hidden and second hidden layer:", mlp.coefs_[1])
print("Weights between hidden and output layer:", mlp.coefs_[2])

# View the biases (intercepts) of the model
print("Biases of the hidden layer:", mlp.intercepts_[0])
print("Biases of the second hidden layer:", mlp.intercepts_[1])
print("Biases of the output layer:", mlp.intercepts_[2])
```
